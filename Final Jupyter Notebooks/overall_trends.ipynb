```python
import sys
print(sys.version)
import numpy as np
print(np.__version__)
import pandas as pd
print(pd.__version__)
import matplotlib.pyplot as plt
import json
import seaborn as sns

%matplotlib inline
```

```python
# Load PIRUS Data
pirus = pd.read_csv('../data/clean_data/pirus_deep_clean_Final.csv')

# Load PIRUS Codebook
with open('../data/clean_data/pirus_codebook.json') as f:
    pirus_codebook = json.load(f)

# Load Custom Color Map
cmap = ['#E7E3DF', '#3186A9', '#EEAD9A', '#DBB03B', '#66759A', '#EBD79F', '#83684C', '#6D6E72', '#965341', '#8F917F', '#010503']
```

```python
# Get Dataframe
instances_per_year = pd.DataFrame(pirus['year'].value_counts()).rename(columns = {'year':'# of Instances'})
instances_per_year.index.name = 'Year'
instances_per_year = instances_per_year.sort_index()
```

```python
# Configure Line Plot
fig, ax = plt.subplots(figsize= [8, 5])
sns.lineplot(x = instances_per_year.index, y = instances_per_year.iloc[:,0], data = instances_per_year, linewidth = 2.5, color = '#3186A9')
ax.set_title('Radicalization has increased from 1980 to 2018', weight='bold', pad = 20, loc='center').set_fontsize('18')
ax.set_ylabel(r"Radical Count", rotation='vertical', size = 12, weight='normal')
ax.set_xlabel("") 
```
